```r
library(ggplot2)
```

```r
my_theme <- theme(axis.title = element_text(size=rel(2)),
             axis.text = element_text(size=rel(1.5)),
             plot.title = element_text(size=rel(2), colour = "blue"),
             panel.grid.major = element_line(colour = "black"))
```

```r
survey <- read.csv(file='archive/LIQuizPollData.csv')
head(survey)
```

```r
ggplot(survey, aes(x = Quiz_number , y = Total_Responses/Total_Views)) + 
        geom_point(size=3)+
        #geom_smooth(method='loess')+
        labs(title='The ratio of responses to survey views') +
        xlab("Quiz number")+
        ylab("Response rate")+
        my_theme
```

Conclusions: Usually over 80% of people who saw a survey turn it off

```r
ggplot(survey, aes(x = Total_Responses , y = Total_Views)) + 
        geom_point() +
        stat_smooth(method='loess')+
        labs(title='The ratio of responses to survey views')+
        xlab("Number of responses")+
        ylab("Number of views")+
        my_theme
```

Conclusion: The more views, the more likes :)

```r
population <- read.csv(file='arab_population_csv.csv')
head(population)
```

```r
ggplot(population, aes(Year, Value)) +
        geom_line() +
        labs(title="Arabian population over the years") +
        xlab("Year") +
        ylab("Population")+
        my_theme +
        theme(panel.grid.major = element_line(colour = "white"))
```

Conclusion: Arabia's population is growing and doing well. The function resembles a square function with a very small slope

```r
bikes <- read.csv(file='bike_buyers/bike_buyers_clean.csv')
head(bikes)
```

```r
ggplot(bikes, aes(Cars, Income), color="purple")+
    stat_bin_hex()+
    scale_fill_gradient(low = "purple", high = "yellow")+
    theme(panel.background = element_rect(fill = "#b3fffa"),
          panel.grid.major = element_line(colour = "black"),
          aspect.ratio=0.95)+
    labs(title="Visualization of customer wealth")+ #depending on income and number of cars owned
    xlab("Cars")+
    ylab("Income")
```

Conclusions: The vast majority of customers have 2 or fewer cars. Most people have 2 cars with an income of about 62.5 thousand (about 80 people). No one with an income over 100,000 owns less than 2 cars. The vast majority of customers with 4 cars have an income of more than 50,000

```r
my_theme_number_2 <- theme(axis.title = element_text(size=rel(2)),
             axis.text = element_text(size=rel(1.5)),
             plot.title = element_text(size=rel(2.5), colour = "blue"),
             panel.grid.major = element_line(colour = "black"),
             legend.background = element_rect(fill="#ececec",
                                              colour="black",
                                              linetype=3),
             legend.title = element_text(size=rel(2)),
             legend.text = element_text(size=rel(1.5))) 
```

```r
ggplot(bikes, aes(Gender))+
        geom_bar(aes(fill=Marital.Status))+
        labs(title="Marital status and gender of clients")+
        xlab("Client gender")+
        ylab("Quantity")+
        scale_fill_discrete(name="Marital status", labels=c("In marriage", "Single"))+
        my_theme_number_2
```

Conclusions: More women are married than men. It can be said that the number of customers is evenly distributed between both sexes
